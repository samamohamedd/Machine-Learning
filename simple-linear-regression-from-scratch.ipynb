```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('california_housing_train.csv')
```

```python
pop = df['population']
brooms = df['total_bedrooms']

plt.scatter(pop, brooms)
plt.xlabel('population')
plt.ylabel('bedrooms')
```

```python
pop_np = pop.to_numpy()
brooms_np = brooms.to_numpy()
```

```python
pop_np.shape, brooms_np.shape
```

```python
from sklearn.linear_model import LinearRegression as LR 

sk_model = LR().fit(pop_np.reshape((17000,1)), brooms_np)
sklearn_bedrooms_predictions = sk_model.predict(pop_np.reshape((17000,1)))
sklearn_bedrooms_predictions.shape
```

```python
pop = df['population']
bedrooms = df['total_bedrooms']

plt.scatter(pop, bedrooms)
plt.xlabel('Population')
plt.ylabel('Total Bedrooms')

plt.scatter(pop, sklearn_bedrooms_predictions)
```

```python
predictions_df = pd.DataFrame({'Population': pop,
                              'bedrooms': bedrooms,
                              'predictions': sklearn_bedrooms_predictions})
predictions_df
```

```python
# the cost function:
def compute_cost(x, y, w, b):
    m = len(y)
    cost = 0

    for i in range(m):
        f_wb = w*x[i] + b 
        cost = cost + (f_wb - y[i])**2

    total_cost = (1/2*m) * cost

    return total_cost
```

```python
# gradient descent function:
def compute_gradien(x, y, w, b):
    m = len(x)   
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = w*x[i] + b
        dj_dw_i = (f_wb - y[i])*x[i]
        dj_db_i = f_wb - y[i]
        dj_dw += dj_dw_i
        dj_dw += dj_db_i

    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_dw, dj_db
```
